```python
import numpy as np
import pandas as pd
```

```python
solar_data = pd.read_csv('data/solarpower.csv')
solar_data.head()
```

```python
for i in range(solar_data['cum_power'].shape[0]):
    if solar_data['cum_power'][i] == -1:
        solar_data['cum_power'][i] = (solar_data['cum_power'][i-1] + solar_data['cum_power'][i+1]) / 2
```

```python
solar_data['cum_power'].mean()
```

```python
solar_data['day_power'] = np.nan
solar_data['day_power'][0] = 0
for i in range(1, solar_data['cum_power'].shape[0]):
     solar_data['day_power'][i] = solar_data['cum_power'][i] - solar_data['cum_power'][i-1]
```

```python
solar_data['month_year_power'] = pd.to_datetime(solar_data['date']).dt.to_period('M')
```

```python
solar_data.groupby('month_year_power')['day_power'].mean()
```

```python
import plotly
import plotly.express as px
```

```python
line_data = solar_data.groupby('month_year_power', as_index=False)['day_power'].mean()
fig = px.line(
    data_frame=line_data,
    x=line_data['month_year_power'].dt.to_timestamp(),
    y='day_power',
    labels={'x':'Год-месяц', 'day_power':'Средняя ежедневная выработка'},
    height=500,
    width=1000,
    title='Средняя ежедневная выработка солнечной энергии по месяцам' 
)
fig.show()
```
